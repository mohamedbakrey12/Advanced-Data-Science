# Apply non-local means denoising

```python


import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, restoration

# Read the image using Matplotlib
img = cv2.imread('/content/Img1.jpeg')

# Convert the image to grayscale
gray = color.rgb2gray(img)

# Apply non-local means denoising
denoised = restoration.denoise_nl_means(gray, h=0.178)  # Adjust h parameter as needed


# Display the original, denoised, and adjusted images using Matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(denoised, cmap='gray')
plt.title('Denoised Image')
plt.axis('off')

plt.show()
```

# Apply median filtering for denoising

```python

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from scipy.ndimage import median_filter

# Read the image using Matplotlib
img = cv2.imread('/content/21.jpeg')

# Convert the image to grayscale
gray = color.rgb2gray(img)

# Apply median filtering for denoising
denoised = median_filter(gray, size=5)  # Adjust the kernel size (5 here) as needed

# Display the original, denoised, and adjusted images using Matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(denoised, cmap='gray')
plt.title('Denoised Image')
plt.axis('off')


plt.show()
```
